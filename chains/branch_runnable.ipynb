>using all runnable to make a    llm to give either output or summary if the  llm output exceed 300 words 

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough,RunnableParallel,RunnableSequence,RunnableBranch
from dotenv import load_dotenv

load_dotenv()
```

```python
model=ChatGroq(model="moonshotai/kimi-k2-instruct-0905")
promt1=PromptTemplate(template="tell me in detail about {topic}",
                      input_variables=["topic"])

prompt2 = PromptTemplate(
    template='Summarize the following text \n {text}',
    input_variables=['text']
)

parsor=StrOutputParser()

def total_length(text):
    return len(text.split())
```

```python
detail = promt1 | model | parsor

branch_chain = RunnableBranch(
    (lambda x: len(x.split())>300, prompt2 | model | parsor),
    RunnablePassthrough()
)

final_chain = RunnableSequence(detail, branch_chain)

print(final_chain.invoke({'topic':'Russia vs Ukraine'}))
```

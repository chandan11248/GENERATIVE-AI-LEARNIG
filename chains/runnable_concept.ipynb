# illustration: how  runnable  make seamless flow in langchain 

```python
from abc import ABC,abstractmethod

class runnable(ABC):
    @classmethod
    def invoke(input_data):
        pass
```

```python
import random
class nakliLLM(runnable):
    def __init__(self):
        print("LLM created")
    def invoke(self,promt):
        response=["kathmandu is the capital of nepal", 
                  "pokhara is one of the beutiful city",
                  "ai is one of the  interesting topic",
                  "NPL is a cricket"]
        return{"content":random.choice(response)}
        


        



        
```

```python
class NAKLIPROMT(runnable):
    def __init__(self,template,input_variable):
        self.input_variable=input_variable
        self.template=template

    def invoke(self,input_variable):
        return self.template.format(**input_variable)
```

```python
class nakliSTRPARSOR(runnable):
    def __init__(self):
        pass         #initial kei chaiyena tara invoke huda input chaiyo

    def invoke(self,input_data):
        return input_data["content"]
    
```

```python
class runnableconnector(runnable):
    def __init__(self,runnable_list):
        self.runnable_list=runnable_list
    def invoke(self,input_data):
        
        for runable in self.runnable_list:
            input_data=runable.invoke(input_data)
        return input_data
    
```

```python
llm=nakliLLM()
template=NAKLIPROMT(template="tell me fact about {country}.",
                    input_variable=["country"])
parsor=nakliSTRPARSOR()

chain=runnableconnector([template,llm,parsor])
chain.invoke({'country':'nepal'})
```

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableParallel,RunnableSequence
from langchain_core.output_parsers import StrOutputParser
```

```python
model=ChatGroq(model="moonshotai/kimi-k2-instruct-0905")
```

```python
promt1=PromptTemplate(template="tell me about pokhara.")
promt2=PromptTemplate(template="tell me about kathmandu")
parsor=StrOutputParser()
```

```python
chain1=promt1|model|parsor
chain2=promt2|model|parsor
```

```python
parallel_Chain=RunnableParallel({
    'pokhara':RunnableSequence(chain1),
    'kathmandu':RunnableSequence(chain2)

})
result=parallel_Chain.invoke({})
print(result)
```

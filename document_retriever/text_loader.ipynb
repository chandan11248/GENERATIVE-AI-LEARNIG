```python
from langchain_community.document_loaders import  DirectoryLoader
loader=DirectoryLoader(path="pdfs",glob="*.pdf"
                       ,loader_cls=PyPDFLoader)
```

```python
docs=loader.load() 
```

```python
print(len(docs)) 
print(docs[0].metadata)
```

```python
docs=loader.lazy_load() #each time meta dat of one document come in memory we print and then remove it then again another comes and so on
for document in docs:
    print(document.metadata)
```

>webloader

```python
from langchain_community.document_loaders import WebBaseLoader
loader=WebBaseLoader("https://www.geeksforgeeks.org/deep-learning/deep-learning-tutorial/")
text=loader.load()
```

```python
len(text[0].page_content)
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv

load_dotenv()
```

```python
model=ChatGroq(model="openai/gpt-oss-120b")
promt=PromptTemplate(template="answer the following question {question} based on given text:{text}",
                      input_variables=["question","text"])

parsor=StrOutputParser()
chain=promt|model|parsor
chain.invoke({'question':'what is deep learning','text':text[0].page_content})
```

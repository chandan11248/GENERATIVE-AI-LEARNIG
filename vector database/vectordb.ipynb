```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_core.documents import Document

# Create LangChain documents for IPL players

doc1 = Document(
        page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and fitness, he has led the Royal Challengers Bangalore in multiple seasons.",
        metadata={"team": "Royal Challengers Bangalore"}
    )
doc2 = Document(
        page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm demeanor and ability to play big innings under pressure.",
        metadata={"team": "Mumbai Indians"}
    )
doc3 = Document(
        page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicketkeeping, and leadership are legendary.",
        metadata={"team": "Chennai Super Kings"}
    )
doc4 = Document(
        page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his yorkers and death-over expertise.",
        metadata={"team": "Mumbai Indians"}
    )
doc5 = Document(
        page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his quick fielding and match-winning performances make him a key player.",
        metadata={"team": "Chennai Super Kings"}
    )
```

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_base=Chroma(embedding_function=embedding,
                   persist_directory='my_chroma_db',
    collection_name='sample')
```

```python
# add documents
vector_base.add_documents(docs)
```

```python
vector_base.get(include=['embeddings','documents', 'metadatas'])
# it has embeddings ,documents and metadatas
```

```python
vector_base.similarity_search(query="who is the bowler",k=3)
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader=WebBaseLoader("https://en.wikipedia.org/wiki/Deep_learning")
text=loader.load()
```

```python
text
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv

load_dotenv()
```

```python
model=ChatGroq(model="openai/gpt-oss-120b")
promt=PromptTemplate(template="answer the following question {question} based on given text:{text}",
                      input_variables=["question","text"])

parsor=StrOutputParser()
chain=promt|model|parsor

necessary_data=Chroma(embedding_function=embedding,
                   persist_directory='sample_db',
    collection_name='NECESSARY_data')
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=700,
                                        chunk_overlap=30,
                                        )
```

```python
chunks=splitter.split_documents(text)
```

```python
len(chunks)
```

```python
necessary_data.add_documents(chunks)
```

```python
similar_splitted_documents=vector_base.similarity_search(query="what is deep learning",k=5)
```

```python
chain.invoke({'question':'what is deep learning','text':similar_splitted_documents})
```
